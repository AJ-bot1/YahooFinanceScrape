```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
pd.set_option('display.max_rows', 100)
```

```python
keys = ['country','industry','sector','marketCap','symbol','exchange','shortName','currency','ask','bid','currentPrice','recommendationKey']
```

# Countries Ranked by MarketCap

```python
url = 'https://companiesmarketcap.com/all-countries/'
page = requests.get(url)
soup = BeautifulSoup(page.content)
```

```python
tab = soup.find('table', class_='default-table table marketcap-table dataTable')
```

```python
urls = ['https://companiesmarketcap.com/' + url['href'] for url in tab.find_all('a', href=True)]
```

```python
urls
```

## Table of TOP10 companies by MarketCap of each countries

```python
top10_tickers = []
all_data = []
for url in urls:
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    tickers = [div.get_text() for div in soup.find_all('div', class_ = 'company-code')]
    for ticker in tickers[:10]:
        try:
            print(ticker)
            yf_tickers = yf.Ticker(ticker)
            info = yf_tickers.get_info()
            data = [info[key] for key in keys]
            all_data.append(data)
            sleep(1)
        except Exception:
            print(f'Skipping ticker: {ticker}...')
```

```python
dtf = pd.DataFrame(all_data, columns = keys)
```
